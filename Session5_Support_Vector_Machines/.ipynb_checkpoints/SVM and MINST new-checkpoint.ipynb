# Visualize and present the dataset

```python
def plot_digits(data):
    fig, axes =plt.subplots (10, 10, figsize=(10, 5),
                            subplot_kw={'xticks':[], 'yticks':[]})
                            #gridspec_kw=dict(hspace=0.1, wspace=0.1))
    for i, ax in enumerate (axes.flat):
        ax.imshow(data[i].reshape(8,8),
                 cmap='binary', interpolation='nearest',
                 clim=(0,16))
        
plot_digits(mnist['data'])
```

```python
from sklearn.manifold import Isomap
iso = Isomap(n_components=10)
projection = iso.fit_transform(mnist.data)
```

```python
plt.scatter(projection[:, 0], projection[:, 1], lw=0.1,
    c=mnist.target, cmap=plt.cm.get_cmap('cubehelix', 6))
plt.colorbar(ticks=range(6), label='digit value')
plt.clim(-0.5, 5.5)
```

# Apply the SVC to perform multi-class classification

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, random_state = 42)
```

```python
sns.countplot(y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=1000)
model.fit(X_train, y_train)
ypred = model.predict(X_test)
```

```python
from sklearn import metrics
print(metrics.classification_report(ypred, y_test))
```

```python
from sklearn.metrics import confusion_matrix
mat = confusion_matrix(y_test, ypred)
sns.heatmap(mat.T, square=True, annot=True, fmt='d', cbar=False)
plt.xlabel('true label')
plt.ylabel('predicted label');
```

```python
from sklearn.svm import SVC
from sklearn import metrics
from sklearn.model_selection import GridSearchCV, ParameterGrid
```

```python
c = [0.0001, 0.001, 0.01, 0.1, 1, 10, 100, 1000]
k = ['linear', 'rbf', 'poly', 'sigmoid']
g = [100, 10, 1, 0.1, 0.01, 0.001, 0.0001]
```

```python
para_grid = dict(C = c,
                kernel =k,
                gamma = g)
```

```python
grid_serch = GridSearchCV(SVC(), param_grid = para_grid, scoring ="accuracy")
```

```python
grid_serch_model = grid_serch.fit(X_train, y_train)
```

```python
CV_res = grid_serch.cv_results_
CV_scores = pd.DataFrame(CV_res["mean_test_score"])
CV_scores.rename (columns ={0:'Avg_Accuracy_Score'}, inplace = True)
CV_res_params =pd.DataFrame(CV_res["params"])
best_scores_SVC =pd.concat([CV_scores, CV_res_params], axis =1)
best_scores_SVC = best_scores_SVC.sort_values(by=['Avg_Accuracy_Score'], ascending=False).head(10)
best_scores_SVC.reset_index().drop(columns=['index'])
```

USING PCA

```python
# A function where we going to test different level of PCA decomposition and see how it improves accuracy
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(2)
projected =pca.fit_transform(mnist.data)
print(mnist.data.shape)
print(projected.shape)
```

```python
plt.scatter(projected[:,0], projected[:,1],
           c=mnist.target,edgecolor='none', alpha=0.5,
           cmap=plt.cm.get_cmap('viridis',20))
plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.colorbar();
```

```python
pca = PCA().fit(mnist.data)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance');
```

```python

def n_component_analysis(n,X_train, y_train, X_test, y_test,kernel_type="rbf"):
   
    pca = PCA(n_components=n)
    print("PCA begin with n_components: {}".format(n))
    pca.fit(X_train)

    X_train_pca = pca.transform(X_train)
    X_test_pca = pca.transform(X_test)

    print('SVC begin')
    clf1 = SVC(kernel=kernel_type)
    clf1.fit(X_train_pca, y_train)
    predictions=clf1.predict(X_test_pca)
    accuracy = metrics.accuracy_score(y_test,predictions)
    
    print("accuracy: {}".format(accuracy))
    return accuracy
```

```python
n_s = np.linspace(0.70, 0.85, num=15)
accuracy = []
for n in n_s:
    tmp = n_component_analysis(n,X_train, y_train, X_test, y_test,'linear')
    accuracy.append(tmp)
```

```python
del accuracy
accuracy = []
for n in n_s:
    tmp = n_component_analysis(n,X_train, y_train, X_test, y_test,"poly")
    accuracy.append(tmp)
```

```python
import matplotlib.pyplot as plt
plt.figure()
plt.plot(n_s,np.array(accuracy),label='Accuracy vs % variance explained')
plt.legend()
plt.show()
```
